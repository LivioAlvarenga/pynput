## Detectar movimento do mouse

<https://nitratine.net/blog/post/how-to-get-mouse-clicks-with-python/>

```python
from pynput import mouse

def on_move(x, y): # movimentando
    print('Pointer moved to {0}'.format((x, y)))

def on_click(x, y, button, pressed): # ao clicar
    print('{0} at {1}'.format('Pressed' if pressed else 'Released',(x, y)))
    
    # finaliza o scrip por retornar False
    if not pressed: #se comando pressed = False finaliza o monitor
        print('Fim da execução monitor de mouse por return False')  
        return False

def on_scroll(x, y, dx, dy): # scrow dx horizontal e dy vertical
    print('Scrolled {0} at {1}'.format('down' if dy < 0 else 'up',(x, y)))
    # ou finaliza o scrip por retornar listener.stop()
    print('Fim da execução monitor de mouse por listener.stop()')
    listener.stop()
```

```python
# desta forma ligamos o monitor de mouse e entramos em um loop ate que saimos do monitor de mouse 
# veja que o print('Fim !!!') é somente no final
with mouse.Listener(on_move=on_move, on_click=on_click, on_scroll=on_scroll) as listener:
    listener.join()

print('Fim !!!')
```

```python
# desta forma ligamos o monitor de mouse e seguimos o script, veja que o print('Fim !!!') é o primeiro retorno
listener = mouse.Listener(on_move=on_move, on_click=on_click, on_scroll=on_scroll)
listener.start()

print('Fim !!!')
```

### Detectar movimento do mouse com log

```python
from pynput.mouse import Listener
import logging

logging.basicConfig(filename="mouse_log.txt", level=logging.DEBUG, format='%(asctime)s: %(message)s')

def on_move(x, y):
    logging.info("Mouse moved to ({0}, {1})".format(x, y))

def on_click(x, y, button, pressed):
    if pressed:
        logging.info('Mouse clicked at ({0}, {1}) with {2}'.format(x, y, button))

def on_scroll(x, y, dx, dy):
    logging.info('Mouse scrolled at ({0}, {1})({2}, {3})'.format(x, y, dx, dy))
    
    print('Fim da execução monitor de mouse por listener.stop()')
    listener.stop()

with Listener(on_move=on_move, on_click=on_click, on_scroll=on_scroll) as listener:
    listener.join()
```

### Tornando as coisas um pouco mais legíveis

```python
from pynput import keyboard

filename = "key_log.txt"  # The file to write characters to

def on_press(key):
    f = open(filename, 'a')  # Open the file

    if hasattr(key, 'char'):  # Write the character pressed if available
        f.write(key.char)
    elif key == Key.space:  # If space was pressed, write a space
        f.write(' ')
    elif key == Key.enter:  # If enter was pressed, write a new line
        f.write('\n')
    elif key == Key.tab:  # If tab was pressed, write a tab
        f.write('\t')
    else:  # If anything else was pressed, write [<key_name>]
        f.write('[' + key.name + ']')

    f.close()  # Close the file
    
def on_release(key):  # The function that's called when a key is released
    
    if key == keyboard.Key.esc: # se pressionar tecla esc
        print('Fim de execução')
        listener.stop() # pode ser por listener.stop()
        #return False # pode ser por return False

with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:  # Setup the listener
    listener.join()  # Join the thread to the main thread
```

# Usando mouse e teclato juntos

<https://nitratine.net/blog/post/how-to-use-pynputs-mouse-and-keyboard-listener-at-the-same-time/>

```python
import pynput as pyp

def on_press(key):
    print("Key pressed: {0}".format(key))

def on_release(key):
    print("Key released: {0}".format(key))
    
    if key == pyp.keyboard.Key.esc: # se teclar esc
        gravador_mouse.stop() # fim mouse
        gravador_teclado.stop() # fim teclado
        print('Fim de execução')

def on_move(x, y): # se quiser ver a movimentação do mouse e so habilitar esta função
    #print("Mouse moved to ({0}, {1})".format(x, y))
    pass

def on_click(x, y, button, pressed):
    if pressed:
        print('Mouse clicked at ({0}, {1}) with {2}'.format(x, y, button))
    else:
        print('Mouse released at ({0}, {1}) with {2}'.format(x, y, button))

def on_scroll(x, y, dx, dy):
    print('Mouse scrolled at ({0}, {1})({2}, {3})'.format(x, y, dx, dy))
    
# Setup the listener threads
gravador_teclado = pyp.keyboard.Listener(on_press=on_press, on_release=on_release)
gravador_mouse = pyp.mouse.Listener(on_move=on_move, on_click=on_click, on_scroll=on_scroll)

# Start the threads and join them so the script doesn't end early
gravador_teclado.start()
gravador_mouse.start() 
gravador_teclado.join()
gravador_mouse.join()

print('FIM Codigo')
```

## Monitorador de mouse

```python
from time import time
from pynput import mouse

def duplo_click(inicio, evento):
    """Função criada para detectar duplo clique de mouse:
    mede-se o tempo ao soltar o primeiro clique ate ao soltar o segundo clique,
    se tempo < que 0.3 segundos é duplo clique devolvendo um True.

    Args:
        inicio (float): tempo inicial medido para comparar como novo tempo (time())
        evento (pynput): evento detectado do mouse da biblioteca pynput

    Returns:
        inicio (float): tempo atual do sistema que foi clicado o botão esquerdo do mouse para fazer nova comparação
        bool: True se duplo clique, False clique comum
    """
    # somente eventos "soltando(pressed=False) botão esquerdo do mouse(button=Button.left)"
    if evento.find('pressed=False') > 0 and evento.find('button=Button.left') > 0: 
        if (time() - inicio) > 0.3: # Tempo agora - tempo que foi clicado o primeiro clique
            inicio = time() # se > 0.3 segundos NÃO É DUPLO CLIQUE, iguala var inicio com tempo de agora para comparar novamente
        else:
            return inicio, True # ativa o duplo clique intervalo entre clique esquerdo < 0.3 segundos
        return inicio, False # clique comum
    return inicio, False # clique comum


def end_mouse(evento):
    """Detecta clique no botão do meio do mouse com objetivo de finalizar o scrip. O scrip é finalizado
    pois retorna um time de 0.01 ou 300 segundos, ou seja,  se o mouse não ter atividade neste tempo 
    finaliza o script.

    Args:
        evento (pynput): evento detectado do mouse da biblioteca pynput

    Returns:
        float: tempo em segundos. 
    """
    # somente eventos "soltando(pressed=False) botão meio do mouse(button=Button.middle)"
    if evento.find('pressed=False') > 0 and evento.find('button=Button.middle') > 0:
        return 0.01
    return 300.0 # 5 minutos sem atividade vai finalizar script


def extrai_comandos(comandos):
    """Recebe uma lista com comandos realizados como texto e prepara uma lista com os padrões 
    de execução da bliblioteca pynput para executar os comandos

    Args:
        comandos (list): lista de comandos executados pelo mouse em str

    Returns:
        lista_comandos (list): lista de comandos limpa e com padrão para execução dos comandos 
        na biblioteca pynput
    """

    lista_comandos = []
    for c in comandos:
        lista_comandos_interna = []
        limpa_texto = c.replace(')','').split(',') # elimina ')' e faz split ','
        for i in range(len(limpa_texto)):
            indice = limpa_texto[i].find('=')
            informacao = limpa_texto[i][indice + 1:]

            if i < 2: # posição 0 e 1 serão coordenadas x e y
                informacao = int(informacao)
            elif i == 3 and informacao =='False': # posição 3 é um bool para cliques mouse
                informacao = bool()
            elif i == 3 and informacao =='True':
                informacao = bool(informacao)
            # posição 2 e 3 é dx e dy para scrow do mouse
            elif (i == 2 or i == 3) and (informacao == '-1' or informacao == '0' or informacao == '1'): # scrow
                informacao = int(informacao)

            lista_comandos_interna.append(informacao)
        lista_comandos_interna_tupla = tuple(lista_comandos_interna)   
        lista_comandos.append(lista_comandos_interna_tupla)
    return lista_comandos
```
